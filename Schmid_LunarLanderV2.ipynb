```python
# Uczenie agenta opartego o algorytm PPO dla środowiska "Lunar-Lander-v2"

import gym
from stable_baselines3 import PPO
from stable_baselines3.common.env_util import make_vec_env


env = make_vec_env("LunarLander-v2", n_envs=8)

model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=100000)
model.save("ppo_lunarlander")
```

```python
# wykorzystanie wrapera do łatwiejszej wizualizacji działana nauczonego agenta - PPO - LunarLander:

import gnwrapper

env = gnwrapper.Monitor(gym.make("LunarLander-v2"),directory="./p3/")

observation = env.reset()
total_reward = 0
while True:
    action, _states = model.predict(observation)
    observation, reward, done, info = env.step(action)
    total_reward += reward
    if done:
        env.reset()
        break

env.display()

print(f"Total reward: {total_reward}")
```
